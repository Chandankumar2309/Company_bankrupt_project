From the above pie chart we can clearly see that the dataset is highly imbalanced

#### As the dataset is already preprocessed at a good level we will keep our focus on fixing the imbalace in the dataset


```python
#splitting the data
from sklearn.model_selection import train_test_split
x,y=df.drop('Bankrupt?',axis=1),df['Bankrupt?']
X_train,X_test,y_train,y_test=train_test_split(x,y,stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
#Method 1
from imblearn.over_sampling import SMOTE
smote=SMOTE()
X_resampled,y_resampled=smote.fit_resample(X_train,y_train)
```

```python
y_resampled.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(class_weight='balanced')
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf1=RandomForestClassifier()
clf1.fit(X_resampled,y_resampled)
clf2=RandomForestClassifier()
clf2.fit(X_train,y_train)
```

```python
clf1.score(X_test,y_test)
```

```python
y_pred1=clf1.predict(X_test)
```

```python
y_pred2=clf2.predict(X_test)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print(classification_report(y_test, y_pred1))
# print("AUC:", roc_auc_score(y_test, y_proba))
```

```python
print(classification_report(y_test, y_pred2))
```

```python
y_pred=clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier

bclf=BalancedRandomForestClassifier()
bclf.fit(X_train,y_train)
y_bpred=bclf.predict(X_test)
```

```python
print(classification_report(y_test, y_bpred))
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_under_resampled, y_under_resampled = rus.fit_resample(X_train, y_train)
```

```python
clf3=RandomForestClassifier()
clf3.fit(X_under_resampled,y_under_resampled)
```

```python
y_u_pred=clf3.predict(X_test)
print(classification_report(y_test, y_u_pred))
```
